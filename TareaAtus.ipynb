```python
import pandas as pd
import matplotlib.pyplot as plt
atus2021 = pd.read_csv("./conjunto_de_datos/atus_anual_2021.csv", index_col=False)
atus2022 = pd.read_csv("./conjunto_de_datos/atus_anual_2022.csv", index_col=False)
atus2023 = pd.read_csv("./conjunto_de_datos/atus_anual_2023.csv", index_col=False)
pd.set_option('display.max_columns', None)
```

# 1. Cargue en un Dataframe los datos de los accidentes de los últimos 3 años.


```python
data = pd.concat([atus2021, atus2022, atus2023], ignore_index=True)
```

```python
data
```

# 2. Agregue una columna tipo date formada por los valores de las columnas correspondientes a Año, Mes, Dia, Hora y Minuto.

```python
data['DATE'] = data.apply(lambda row: f"{row['ANIO']}-{str(row['MES']).zfill(2)}-{str(row['ID_DIA']).zfill(2)} {str(row['ID_HORA']).zfill(2)}:{str(row['ID_MINUTO']).zfill(2)}", axis=1)
data.insert(3, 'DATE', data.pop('DATE'))
```

```python
data
```

# 3. Agregue una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.

```python
def asignar_area(row):
    mapping = {
        'Sin accidente en esta zona': 'SUBURBANA',
        'Accidente en intersección': 'URBANA',
        'Accidente en no intersección': 'URBANA',
        'Accidente en carretera estatal': 'SUBURBANA'
    }
    if row['URBANA'] in mapping:
        return mapping[row['URBANA']]
    elif row['SUBURBANA'] in mapping:
        return mapping[row['SUBURBANA']]
    return None

data['AREA'] = data.apply(asignar_area, axis=1)
```

```python
data
```

# 4. Agregue una columna Zona cuyos valores serán la zona donde ocurrió el accidente.

```python
def asignar_zona(row):
    if row['AREA'] == 'SUBURBANA':
        return row['SUBURBANA']
    elif row['AREA'] == 'URBANA':
        return row['URBANA']
    
    
data["ZONA"] = data.apply(asignar_zona, axis=1)
```

```python
data
```

# 5. Elimine las columnas URBANA y SUBURBANA.

```python
data = data.drop(columns=["URBANA"])
data = data.drop(columns=["SUBURBANA"])
```

# 7. Resuma los accidentes del último año por tipo y que porcentaje representan.

```python
accidentes_2023_2 = data[data["ANIO"] == 2023]
accidentes_2023 = accidentes_2023_2[accidentes_2023_2 ['ESTATUS'] != 'Certificado Cero']
resumen_accidentes = accidentes_2023.groupby('TIPACCID').size().reset_index(name='CANTIDAD')
total_accidentes = resumen_accidentes['CANTIDAD'].sum()
resumen_accidentes['PORCENTAJE'] = (resumen_accidentes['CANTIDAD'] / total_accidentes) * 100
resumen_accidentes
```

# 8. Obtenga la cantidad de muertos y lesionados de los últimos 3 años.

```python
data_2021 = data[data["ANIO"] == 2021]
numMuertos_2021 = data_2021[['CONDMUERTO', 'PASAMUERTO', 'PEATMUERTO', 'CICLMUERTO', 'OTROMUERTO', 'NEMUERTO']].sum().sum()

data_2022 = data[data["ANIO"] == 2022]
numMuertos_2022 = data_2022[['CONDMUERTO', 'PASAMUERTO', 'PEATMUERTO', 'CICLMUERTO', 'OTROMUERTO', 'NEMUERTO']].sum().sum()

data_2023 = data[data["ANIO"] == 2023]
numMuertos_2023 = data_2023[['CONDMUERTO', 'PASAMUERTO', 'PEATMUERTO', 'CICLMUERTO', 'OTROMUERTO', 'NEMUERTO']].sum().sum()

print("Cantidad de muertos en 2021: "+str(numMuertos_2021) + "\n" + "Cantidad de muertos en 2022: "+str(numMuertos_2022) + "\n" + "Cantidad de muertos en 2023: "+str(numMuertos_2023) + "\n" + "Total de muertos en los ultimos 3 años: "+str(numMuertos_2021+numMuertos_2022+numMuertos_2023))
```

```python
dataLesionados_2021 = data[data["ANIO"] == 2021]
numLesionados2021 = dataLesionados_2021[['CONDHERIDO', 'PASAHERIDO', 'PEATHERIDO', 'CICLHERIDO', 'OTROHERIDO', 'NEHERIDO']].sum().sum()

dataLesionados_2022 = data[data["ANIO"] == 2022]
numLesionados2022 = dataLesionados_2022[['CONDHERIDO', 'PASAHERIDO', 'PEATHERIDO', 'CICLHERIDO', 'OTROHERIDO', 'NEHERIDO']].sum().sum()

dataLesionados_2023 = data[data["ANIO"] == 2023]
numLesionados2023 = dataLesionados_2023[['CONDHERIDO', 'PASAHERIDO', 'PEATHERIDO', 'CICLHERIDO', 'OTROHERIDO', 'NEHERIDO']].sum().sum()

print("Cantidad de lesionados en 2021: "+str(numLesionados2021) + "\n" + "Cantidad de lesionados en 2022: "+str(numLesionados2022) + "\n" + "Cantidad de lesionados en 2023: "+str(numLesionados2023) + "\n" + "Total de lesionados en los ultimos 3 años: "+str(numLesionados2021+numLesionados2022+numLesionados2023))

```

# 9. ¿Cómo se compara el número de accidentes de cada año contra el anterior? ¿En qué porcentaje varió?.

```python
accidentes_2021_2 = data[data["ANIO"] == 2021]
accidentes_2021 = accidentes_2021_2[accidentes_2021_2 ['ESTATUS'] != 'Certificado Cero']
numAccidentes_2021 = accidentes_2021.shape[0]

accidentes_2022_2 = data[data["ANIO"] == 2022]
accidentes_2022 = accidentes_2022_2[accidentes_2022_2 ['ESTATUS'] != 'Certificado Cero']
numAccidentes_2022 = accidentes_2022.shape[0]

accidentes_2023_2 = data[data["ANIO"] == 2023]
accidentes_2023 = accidentes_2023_2[accidentes_2023_2 ['ESTATUS'] != 'Certificado Cero']
numAccidentes_2023 = accidentes_2023.shape[0]

variacion_2022 = ((numAccidentes_2022 - numAccidentes_2021) / numAccidentes_2021) * 100
variacion_2023 = ((numAccidentes_2023 - numAccidentes_2022) / numAccidentes_2022) * 100

print("Cantidad de accidentes en 2021: " + str(numAccidentes_2021))
print("Cantidad de accidentes en 2022: " + str(numAccidentes_2022) + " (Variación: {:.2f}%)".format(variacion_2022))
print("Cantidad de accidentes en 2023: " + str(numAccidentes_2023) + " (Variación: {:.2f}%)".format(variacion_2023))
```
